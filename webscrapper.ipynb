```python
import requests
import os
from dotenv import load_dotenv

load_dotenv()

key = os.environ.get("ACCESS_TOKEN")

BASE_URL = "https://api.genius.com"
params={
    'per_page': 50,
    "page": 1
}
headers = {'Authorization': f"Bearer {key}"}
search_term = "355608"
search_url = f"{BASE_URL}/artists/{search_term}/songs"

eden = {}
while True:
    response = requests.get(search_url, headers=headers, params=params)
    if response.status_code == 200:
        data = response.json()
        songs = response.json()['response']['songs']
        if len(songs) == 0:
            break
        else:
            for song in data['response']['songs']:
                eden[song['title']] = song['relationships_index_url']
            params['page'] = params['page'] + 1
    else:
        print(f"Error: {response.status_code}")
    
print(eden)
```

```python
for _, url in eden.items(): 
    soup = BeautifulSoup(requests.get(url, headers=headers).text, 'html.parser')
    lyrics_div = soup.select_one("div[data-lyrics-container='true']")

    if lyrics_div:
        lyrics_text = lyrics_div.get_text(separator="\n")
        print(lyrics_text)
    else:
        print("Could not find the lyrics container.")
        lyrics_text = None
```
